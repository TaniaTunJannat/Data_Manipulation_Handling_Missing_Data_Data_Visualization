**8. Cluster the final marks of each student using K-Means clustering algorithm. (1<K<6)
Visualize the final clusters.**

```python
df=pd.read_csv("final result.csv")
```

```python
df
```

```python
#Extract the final marks column for clustering and 2D array
X=df[['Final Marks']].values
```

```python
#Determine the optimal number of clusters using the Elbow Method
inertia=[]
K_values=range(2, 6)
for k in K_values:
    kmeans=KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)
```

```python
#Plot the Elbow Method graph
plt.figure(figsize=(6, 4))
plt.plot(K_values, inertia, marker='o', linestyle='-')
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.title("Elbow Method")
plt.show()
```

```python
#The elbow method shows that 3 is a good value for K, so we retrain and visualize the result
best_k=3
kmeans=KMeans(n_clusters=best_k)
df['Cluster']=kmeans.fit_predict(X)
#Improved visualization for 1D clustering
plt.figure(figsize=(8, 5))
plt.scatter(X, np.zeros_like(X), c=df['Cluster'],s=100)
#Mark centroids
centroids = kmeans.cluster_centers_
plt.scatter(centroids, np.zeros_like(centroids), c='red', s=200, marker='X', label='Centroids')
plt.title(f"K-Means Clustering (K={best_k})")
plt.xlabel("Final Marks")
plt.ylabel("Clusters")
plt.show()
```
